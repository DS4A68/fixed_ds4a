```python
df_time = data[['FECHA HECHO','CANTIDAD']]
df_time['FECHA HECHO'] = pd.to_datetime(df_time['FECHA HECHO'])
df_time  = df_time.groupby('FECHA HECHO')['CANTIDAD'].sum().to_frame().reset_index()
df_time.to_csv('df_time_all.csv')


#fig = px.line(df_time, x  ='FECHA HECHO',y = 'CANTIDAD')
```
